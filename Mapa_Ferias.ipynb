Una vez más, descargamos y mapeamos las ferias libres de Santiago

```python
import pandas as pd
#url = 'https://tinyurl.com/feriaslibrespub'  # raw
url = 'https://raw.githubusercontent.com/sergiolucero/data/master/FeriasLibres.csv'
df = pd.read_csv(url)
df.sample(3)
```

```python
df.Comuna.value_counts()
```

```python
alto = df[df.Comuna.isin(['Las Condes','La Reina','Vitacura'])].dropna()
alto
```

```python
import folium
icon = 'info-sign'
hstr = lambda row: pd.DataFrame(row).to_html()
fm = folium.Map(location=[-33.4, -70.55], zoom_start=12,
                width=600, height=400, tiles='Stamen Toner')
for _, row in alto.iterrows():
  folium.Marker(location=[row['Latitud'],row['Longitud']], popup=hstr(row),
                icon=folium.Icon(color='red',icon=icon)).add_to(fm)
fm
```

```python
df.to_csv('FeriasLibres.csv',index=False)
```
