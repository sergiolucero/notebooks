```python
import pandas as pd
url = 'https://dashboards.sdgindex.org/static/downloads/files/SDR%202021%20-%20Database.xlsx'
df = pd.read_excel(url, sheet_name='Raw Data')
df.head()   # muestra los 5 primeros paÃ­ses
```

```python
# First 10
cols = [col for ix, col in enumerate(df.columns) if ix in (6,10,16)]
cols
```

```python
odf = df[df['Regions used for the SDG Index & Dashboard']=='OECD']
cdf = odf[cols]
cdf.index = odf.Country
import seaborn as sns
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1, figsize=(20,12)); 
sns.heatmap(cdf, cmap='jet', annot=True, annot_kws={'size':14, 'weight':'bold'}); 
plt.xticks(rotation=0)
plt.title('SDR 2021 para paÃ­ses OCDE', size=18);
plt.savefig('SDG_plot2.png')
```
